# Introduction to Pandas DataFrames

This notebook demonstrates basic pandas DataFrame operations using Python and Jupyter Notebook.

The examples include:

- Creating a DataFrame
- Viewing rows
- Displaying information
- Generating statistics
- Selecting columns

## Pandas Library

`pandas` is a Python library used for data analysis and data manipulation.

```python
# Python version is 3.13.9

import pandas as pd

data = {
    "Name": ["Ali", "Sara", "John", "Mina"],
    "Age": [25, 30, 22, 28],
    "City": ["Baghdad", "Erbil", "London", "Berlin"]
}

df = pd.DataFrame(data)

df
```

## Displaying the First Rows

The head() function displays the first rows of the DataFrame.

```python
df.head(2)
```

## DataFrame Information

The info() function provides information about the structure and data types.

```python
df.info()
```

## Statistical Summary

The describe() function generates descriptive statistics.

```python
df.describe()
```

```python
df.describe()
```

## Selecting a Column

Specific columns can be selected using square brackets.

```python
df["Name"]
```
